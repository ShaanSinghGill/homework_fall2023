```python
import tensorflow as tf
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
import numpy as np
```

```python
log_file = 'data/hw3_dqn_dqn_CartPole-v1_s64_l2_d0.99_04-10-2023_23-25-46' # Replace with the path to your event log directory
event_acc = event_accumulator.EventAccumulator(log_file)
event_acc.Reload()
tags = event_acc.Tags()
evals = [v.value for v in event_acc.Scalars('eval_return')]
steps = [v.step for v in event_acc.Scalars('eval_return')]
plt.figure(figsize=(10, 5))
plt.plot(steps, evals, label='cartPole')
plt.xlabel('eval_ep_len')
plt.ylabel('eval_return')
plt.title('eval_return vs step')
plt.legend()
plt.grid(True)
plt.show()
```
