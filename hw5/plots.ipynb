```python
import tensorflow as tf
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
import numpy as np
```

```python
log_file = 'data/alpha/hw5_offline_PointmassMedium-v0_cql0_15-11-2023_11-02-34'
event_acc = event_accumulator.EventAccumulator(log_file)
event_acc.Reload()
tags = event_acc.Tags()
print(tags)
evals1, steps1 = [v.value for v in event_acc.Scalars('eval_return')], [
    v.step for v in event_acc.Scalars('eval_return')]

log_file = 'data/hw5_offline_PointmassMedium-v0_cql5_28-11-2023_15-22-22'
event_acc = event_accumulator.EventAccumulator(log_file)
event_acc.Reload()
tags = event_acc.Tags()
print(tags)
evals05, steps05 = [v.value for v in event_acc.Scalars('eval_return')], [
    v.step for v in event_acc.Scalars('eval_return')]


log_file = 'data/alpha/hw5_offline_PointmassMedium-v0_cql10_15-11-2023_11-05-33'
event_acc = event_accumulator.EventAccumulator(log_file)
event_acc.Reload()
tags = event_acc.Tags()
evals2, steps2 = [v.value for v in event_acc.Scalars(
    'eval_return')], [v.step for v in event_acc.Scalars('eval_return')]

log_file = 'data/cqlNdqn/hw5_offline_PointmassMedium-v0_cql0.1_15-11-2023_09-44-22'
event_acc = event_accumulator.EventAccumulator(log_file)
event_acc.Reload()
tags = event_acc.Tags()
evals3, steps3 = [v.value for v in event_acc.Scalars(
    'eval_return')], [v.step for v in event_acc.Scalars('eval_return')]

log_file = 'data/hw5_offline_PointmassMedium-v0_cql1_28-11-2023_15-48-15'
event_acc = event_accumulator.EventAccumulator(log_file)
event_acc.Reload()
tags = event_acc.Tags()
evals4, steps4 = [v.value for v in event_acc.Scalars(
    'eval_return')], [v.step for v in event_acc.Scalars('eval_return')]



plt.figure(figsize=(10, 5))
plt.plot(steps1, evals1, label='Alpha 0')
plt.plot(steps3, evals3, label='Alpha 0.1')
plt.plot(steps4, evals4, label='Alpha 1')
plt.plot(steps05, evals05, label='Alpha 5')
plt.plot(steps2, evals2, label='Alpha 10')

plt.xlabel('epoch')
plt.ylabel('eval_return')
plt.title('CQL eval_return vs step')
plt.legend()
plt.grid(True)
plt.show()
```
